```python
browser_test=webdriver.Firefox(path_tounzipfile)
def get_data_links(url_test):
    browser_test.get(url_test)
    html_test=browser_test.page_source
    soup_test=BeautifulSoup(html_test,'html.parser')
    
    date1=[]
    type_terror1=[]
    dead1=[]
    injured1=[]
    location1=[]
    details1=[]
    perpetrator1=[]
    part_of1=[]
    
    tab_test=soup_test.find_all('table',class_='wikitable')
    rows_terror1=tab_test[0].find_all('tr')
    
    for i in rows_terror1[1:]:
        date1.append(i.find_all('td')[0].text.strip())
        type_terror1.append(i.find_all('td')[1].text.strip())
        dead1.append(i.find_all('td')[2].text.strip())
        injured1.append(i.find_all('td')[3].text.strip())
        location1.append(i.find_all('td')[4].text.strip())
        details1.append(i.find_all('td')[5].text.strip())
        perpetrator1.append(i.find_all('td')[6].text.strip())
        try:
            part_of1.append(i.find_all('td')[7].text.strip())
        except:
            part_of1.append("Data not available")
        
    data_frame2=pd.DataFrame({'date':date1,'type_terror':type_terror1,'dead':dead1,'injured':injured1,'location':location1,
                                  'details':details1,'perpetrator':perpetrator1,'part_of':part_of1})
    data_frame2['url']=url_test
    
    return data_frame2

#url_t='https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_August_2019'
#get_data_links(url_t)


base_url='https://en.wikipedia.org{}'
master={}
for i in linked_list:
    url_test1=i
    master[i]=get_data_links(url_test1)
```

```python
total=pd.concat([master[i] for i in master],axis=0)
total.head()
```

```python
store_year=[]
store_month=[]
for i in total['url']:
    store_year.append(i.split('_')[-1])
    store_month.append(i.split('_')[-2])
store_year
```

```python
#By month of all the years, what is the average no. of dead or injured.

#Cleaning dead data:
import re
#total['dead'].unique()
pattern=re.compile('\d+')
def get_numbers(x):
    match=re.findall(pattern,x)
    if len(match)==0:
        return np.nan
    elif len(match)>1:
        return float(match[-1])
    else:
        return float(match[0])
    
        
total['dead_cleansed']=total['dead'].map(get_numbers)
total.head()
```

```python
total.groupby(['month','year'])['dead_cleansed'].mean()
```
