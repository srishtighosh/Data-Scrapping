### Question 2:
Your next task is to put the data that youve obtained by making an API call into a csv file. Following attributes should be
the columns in the csv file: 

```python
#(i)Latitude (Hint: Treat, the first entry in coordinates attribute as Lat)

new_lattitude=[]
for i in response['features']:
    new_lattitude.append(i['geometry']['coordinates'][0])
new_lattitude
```

```python
#(ii)Longitude (Hint: Treat, the second entry in the coordinates attribute as Longitude)

new_longitude=[]
for i in response['features']:
    new_longitude.append(i['geometry']['coordinates'][1])
new_longitude
```

```python
#(iii)Title : This should include the Earthquake description
response['metadata']['title']
title=[]
for i in response['features']:
    title.append(i['properties']['title'])
title
```

```python
#(iv)Place: The location of the Earthquake

place=[]
for i in response['features']:
    place.append(i['properties']['place'])
place
```

```python
#(v)Mag: Magnitude of the earthquake

magnitude=[]
for i in response['features']:
    magnitude.append(i['properties']['mag'])
magnitude
```

```python
table=pd.DataFrame({'Latitude':new_lattitude,'Longitude':new_longitude,'Title':title,'Place':place,'Magnitude':magnitude})
table.head()
```

```python
table.to_csv('New_Data.csv')
```
